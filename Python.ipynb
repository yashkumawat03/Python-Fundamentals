## **Sets**

Sets are the built in data types in python used to strore collections of unique items. They are unordered,, and they donot allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries and performing mathematical set operations like union, intersection, difference and symmetric difference.

````python
## Mathematical Operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## union
union_set = set1.union(set2)
print(union_set)

## intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

## difference
difference_set = set1.difference(set2)
print(difference_set)

## intersection_update
set1.intersection_update(set2)
print(set1)

## differnece_update
set2.difference_update(set1)
set2

## symmetric_difference
set1.symmetric_difference(set2)

set1.symmetric_difference_update(set2)
set1
````

````python
## Temperature Conversion

def temp_conversion(temp,unit):
  if unit == 'C':  ## Celcius to Farenheit
    return temp* 9/5 + 32
  elif unit == 'F':  ## Farenheit to Celsius
    return (temp-32)*5/9

print(temp_conversion(77,'F'))
print(temp_conversion(25,'C'))
````

````python
## Password Strength Checker
def is_strong_password(password):
  """Checks whether the password is string or not"""
  if len(password)<8:
    return False
  if not any(char.isdigit() for char in password):
    return False
  if not any(char.islower() for char in password):
    return False
  if not any(char.isupper() for char in password):
    return False
  if not any(char in '!@#$%&*()_+' for char in password):
    return False
  return True
## Calling Function
print(is_strong_password("12345678"))
print(is_strong_password("1234Yash"))
print(is_strong_password("Yash@567"))
````

## **Lambda Functions**

Lambda Functions are small anonymous functions(a function without a name) defined using the lambda keyword.
They can have any number of arguments but only one expression.
they are commonly used for short operations or as arguments to highere-order functions

````python
## Syntax
lambda argument : expression
````

## **map()**

The map() function applies a given function to all itmes in an input list(or any other iterable) and return a map object(an iterator). This is particularly useful for transforming data in a list comprehensively.




````python
def square(num):
  return num**2
square(23)
````

````python
numbers = [1,2,3,4,5,6,7,8]
## map(function,iterable)
list(map(square,numbers))
````

````python
## Lambda function with map
numbers = [1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
````

````python
## Map multiple iterables

number1=[1,2,3]
number2=[4,5,6]

added_numbers = list(map(lambda x,y:x+y,number1,number2))
print(added_numbers)
````

````python
## map() to convert a list of string to integers
# use map to convert string to integers

s = ['1','2','3','4','5']
int_numbers=list(map(int,s))
print(int_numbers)
````

````python
words = ['apple','banana','cherry']
upper_words = list(map(str.upper,words))
print(upper_words)
````
